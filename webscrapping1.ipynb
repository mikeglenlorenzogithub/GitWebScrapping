```python
from urllib.request import urlopen
keyword = "gaming"
html = urlopen("https://datascripmall.id/catalogsearch/result/?q="+keyword).read()
print(type(html))
print(html[1:1000])
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(html, "lxml")
print(type(soup))
print(soup.prettify()[1:1000])
```

```python
product = soup.find_all("li","item product product-item")

for p in product:
    print(p.find("a", "product-item-link").get_text())
    print(p.find("span", "price").get_text().replace("Rp","").replace(".",""))
```

```python
import pandas as pd

product_name = []
product_price = []

product = soup.find_all("li","item product product-item")

for p in product:
    name = p.find("a", "product-item-link").get_text().replace("\n","").encode("utf-8")
    price = p.find("span", "price").get_text().replace("Rp","").replace(".","").encode("utf-8")
    
    product_name.append(name)
    product_price.append(int(price))

product_dict = {"Product": product_name, "Price": product_price}
df = pd.DataFrame(product_dict, columns = ["Product","Price"])

df
```

```python
df.to_csv("output_page_1.csv", sep=",")
```
